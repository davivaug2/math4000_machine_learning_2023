### Problem 1

#### Solution to part 1

```python
import matplotlib.pyplot as plt
import numpy as np

def ReLU(x):
    return np.maximum(x,0)

def FNN(x):
    
    W1 = np.array([[1],[1]])
    b1 = np.array([[-1],[0]])
    W2 = np.array([[1, -1],[1, 2]])
    b2 = np.array([[0],[0]])
    W3 = np.array([2, 1])
    b3 = np.array([0])

    z1 = x
    z2 = ReLU(W1*x + b1)
    z3 = ReLU(W2.dot(z2) + b2)
    y = W3.dot(z3) + b3
    
    return y

x = np.linspace(-3, 3, 100)
y = FNN(x)

plt.figure()
plt.plot(x,y)
```

#### Solution to part 2

```python
x = np.array([1])
y = FNN(x)

print(y[0])
```

### Problem 2
- The number of parameters in the first hidden layer is: $H + H$

- The number of parameters in the following $(L-1)$ hidden layer is: $(L-1)  (H^2 + H)$

- The number of parameters in the output layer is: $H + 1$

Therefore, the total number of parameters is $H + H + (L-1)(H^2 + H) + H + 1 = (L-1)H^2 + (L+2)H + 1$

#### Solution to part 2
- The local minimum is $x=-1$ and $x=2$
- The global minimum is $x=2$

#### Solution to part 3

```python
x = -2
eta = 0.001

N = 1000
for i in range(N):
    x = x - eta * (12*x**3 - 12*x**2 - 24*x)
print(x)

# x = -2 and eta = 0.001 will result in the local minimum instead of global minimum. The gradient descent may get stuck in the local minimum if the initial guess is too far away from the global minimum.
# x = 1 and eta = 0.001 will result in the correct global minimum
# x = 1 and eta = 0.1 will result in overflow. The learning rate is too large
```

#### Solution to part 2

```python
import numpy as np

x, y = 1, 1     # initial guess
eta = 1e-2      # learning rate
N = 1000        # number of iterations

for i in range(N):
    x = x - eta * 2*x
    y = y - eta * 2*y
    
print("x = ", x)
print("y = ", y)
```
