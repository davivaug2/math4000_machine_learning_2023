### neural network with one hidden layer

```python
import matplotlib.pyplot as plt
import numpy as np

def ReLU(x):
    return np.maximum(x,0)

x = np.linspace(-2, 2, 100)
y = ReLU(2*x-2) + ReLU(x)

plt.figure()
plt.plot(x,y)
```

### gradient descent

```python
# example, gradient descent in 1D
# minimize function f(x) = x^4

x = 1
eta = 0.05
N = 1000
for i in range(N):
    x = x - eta * 4 * x**3
    print(i, x)
```
