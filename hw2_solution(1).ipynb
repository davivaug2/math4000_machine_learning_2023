```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator

# Problem 2

fig, ax = plt.subplots(subplot_kw={"projection": "3d"})

N = 20
x1 = np.random.rand(N)
x2 = np.random.rand(N)
y = 2*x1 + 3*x2 + 1 + np.random.rand(N) * 0.2

X = np.stack((x1, x2, np.ones(N)), axis=-1)
Y = np.transpose(y)

A = np.matmul(np.transpose(X), X)
coeff = np.matmul(np.linalg.inv(A), np.matmul(np.transpose(X), Y))
m1, m2, b = coeff[0], coeff[1], coeff[2]
print('m1: ', m1)
print('m1: ', m1)
print('m1: ', m1)

ax.scatter(x1, x2, y)

x1, x2 = np.meshgrid(x1, x2)
surf = ax.plot_surface(x1, x2, m1*x1 + m2*x2 + b, cmap=cm.coolwarm, linewidth=0, antialiased=False, alpha = 0.05)

# Customize the z axis.
ax.set_zlim(0, 10)
ax.zaxis.set_major_locator(LinearLocator(10))
# A StrMethodFormatter is used automatically
ax.zaxis.set_major_formatter('{x:.02f}')

# Add a color bar which maps values to colors.
fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```
