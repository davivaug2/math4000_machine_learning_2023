### Problem 1

Read the code for using Euler method to solve ODE. Modify the code to solve the following ODE system for $0\le t \le 1$:
$$
\begin{aligned}
\frac{dx}{dt} &= x \\
\frac{dy}{dt} &= -x + 2y
\end{aligned}
$$
with the initial condition
$$
x(0) = -1, \quad y(0) = 1.
$$
In the figure, compare the solution generated by Euler method with the exact solution:
$$
x(t) = -e^t, \quad y(t) = -e^t + 2 e^{2t}.
$$
