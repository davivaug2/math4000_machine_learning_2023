### linear regression

```python
import matplotlib.pyplot as plt
import numpy as np

N = 10
x = np.linspace(-1, 1, N)
# print(x)
y = 2 * x + 1 + np.random.rand(N) * 0.8
print(y)
plt.figure()
plt.plot(x,y,'o')
```

```python
m = (N*np.sum(x*y) - np.sum(x)*np.sum(y))/(N*np.sum(x*x) - (np.sum(x))**2)
b = (np.sum(y) - m*np.sum(x))/N
test = 4*N-np.sum(x*x)-(np.sum(x))**2
print(test)
print(m, b)

plt.figure()
plt.plot(x,y,'o')
plt.plot(x,m*x+b,'-')
```

### activation function

```python
import matplotlib.pyplot as plt
import numpy as np

# sigmoid function
x = np.linspace(-10, 10, 100)
y = 1/(1+np.exp(-x))

plt.figure()
plt.plot(x,y)
```
